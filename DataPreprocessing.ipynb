## IMDB dataset

```python
%%capture
imdb = load_dataset('imdb', split='train')['text']
```

```python
len(imdb)
```

```python
# remove undesired tokens
ds = [i.replace('<br />', ' ') for i in imdb]
ds = [re.sub(r'[^a-zA-Z\d\s:\.\!\?\']','',i) for i in ds] 
```

```python
# Store data (serialize)
with open('/content/drive/MyDrive/InforetProject/Datasets/imdb_dataset.pickle', 'wb') as handle:
    pickle.dump(ds, handle, protocol=pickle.HIGHEST_PROTOCOL)

# Load data (deserialize)
with open('/content/drive/MyDrive/InforetProject/Datasets/imdb_dataset.pickle', 'rb') as handle:
    unserialized_data = pickle.load(handle)

print(ds == unserialized_data)
```

## Entails Dataset

```python
path = "/content/drive/MyDrive/InforetProject/Datasets/ENTAILMENT_GRAPH_ENG_V2.0/data/ENTAILMENT_GRAPH_RTE_FORMAT_ENG_V2.0/rteFormatData"
```

```python
files = [f for f in listdir(path)]
trees = []

pairs = []

for file in files:
    trees.append(ET.parse(path+'/'+file))

for tree in trees:
    root = tree.getroot()
    for pair in root.findall('pair'):
        if pair.attrib['entailment'] == 'ENTAILMENT':
            # Little fix
            t = pair.find('t').text.replace("comfortableGenerally", "comfortable")
            h = pair.find('h').text.replace("comfortableGenerally", "comfortable")
            pairs.append((t, h))
```

```python
# Store data (serialize)
with open('/content/drive/MyDrive/InforetProject/Datasets/entails_dataset.pickle', 'wb') as handle:
    pickle.dump(pairs, handle, protocol=pickle.HIGHEST_PROTOCOL)

# Load data (deserialize)
with open('/content/drive/MyDrive/InforetProject/Datasets/entails_dataset.pickle', 'rb') as handle:
    unserialized_data = pickle.load(handle)

print(pairs == unserialized_data)
```
